```python
import os
from sqlalchemy import create_engine
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import sessionmaker
import pandas as pd
```

```python
db_fullpath = 'science-data/mpup.db'

SQLALCHEMY_DATABASE_URL='sqlite:///{}'.format(db_fullpath)
engine = create_engine(
    SQLALCHEMY_DATABASE_URL, connect_args={"check_same_thread": False}
)

SessionLocal = sessionmaker(autocommit=False, autoflush=False, bind=engine)

Base = declarative_base()
```

```python
# City of the provider 
sql = text(
     "SELECT CityoftheProvider"
      "     , sum(NumberofServices), avg(NumberofServices), min(NumberofServices), max(NumberofServices)"
      "  FROM mytable"
      " WHERE CityoftheProvider = :r"
      " GROUP BY CityoftheProvider"
     )
conn.execute(sql, r='ZUNI').fetchall()
```

```python
# City of the provider 
sql = text(
     "SELECT"
     " CountryCodeoftheProvider, StateCodeoftheProvider, CityoftheProvider, ZipCodeoftheProvider"
     "     , sum(NumberofServices), avg(NumberofServices), min(NumberofServices), max(NumberofServices)"
     "  FROM mytable"
     " WHERE CountryCodeoftheProvider = :r"
     " GROUP BY CountryCodeoftheProvider, StateCodeoftheProvider, CityoftheProvider, ZipCodeoftheProvider"
     )
result = conn.execute(sql, r='US').fetchall()


# CountryCodeoftheProvider
# StateCodeoftheProvider
# CityoftheProvider
# ZipCodeoftheProvider
```

```python
sql = text(
     "SELECT"
     " CountryCodeoftheProvider, StateCodeoftheProvider, CityoftheProvider, ZipCodeoftheProvider"
     "     , NumberofServices, NumberofMedicareBeneficiaries, NumberofDistinctMedicareBeneficiaryPerDayServices"
     "     , AverageMedicareAllowedAmount, AverageSubmittedChargeAmount, AverageMedicarePaymentAmount, AverageMedicareStandardizedAmount"
     "  FROM mytable"
     " WHERE (CountryCodeoftheProvider = :c OR :c = '')"
     " AND (StateCodeoftheProvider = :s OR :s = '')"
     " AND (CityoftheProvider = :y OR :y = '')"
     " AND (ZipCodeoftheProvider = :z OR :z = '')"
     )
result = conn.execute(sql, c='US', s='AA', y='APO',  z='39530').fetchall()
#result = conn.execute(sql, c='AR', s='', y='',  z='').fetchall()

# CountryCodeoftheProvider
# StateCodeoftheProvider
# CityoftheProvider
# ZipCodeoftheProvider
```

```python
df = pd.DataFrame(result, columns=["CountryCodeoftheProvider", "StateCodeoftheProvider", "CityoftheProvider", "ZipCodeoftheProvider", "NumberofServices", "NumberofMedicareBeneficiaries", "NumberofDistinctMedicareBeneficiaryPerDayServices", "AverageMedicareAllowedAmount", "AverageSubmittedChargeAmount", "AverageMedicarePaymentAmount", "AverageMedicareStandardizedAmount"]) 
df['NumberofServices'] = df['NumberofServices'].astype(int)
df['NumberofMedicareBeneficiaries'] = df['NumberofMedicareBeneficiaries']
df['NumberofDistinctMedicareBeneficiaryPerDayServices'] = df['NumberofDistinctMedicareBeneficiaryPerDayServices']
df['AverageMedicareAllowedAmount'] = df['AverageMedicareAllowedAmount']
df['AverageSubmittedChargeAmount'] = df['AverageSubmittedChargeAmount']
df['AverageMedicarePaymentAmount'] = df['AverageMedicarePaymentAmount']
df['AverageMedicareStandardizedAmount'] = df['AverageMedicareStandardizedAmount']
df
```
